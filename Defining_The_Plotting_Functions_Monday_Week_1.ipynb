#A PLOT OF THE PION ENERGY NEEDED TO PRODUCE A NEUTRINO OF ENERGY E_v AT VARIOUS ANGLES


```python
E_v = 0         # in function

E_pi_1_complex = 0      # in function
E_pi_1_real = 0         # in function

E_pi_2_complex = 0      # in function
E_pi_2_real = 0         # in function

theta = 0.03

m_pi = 139.6 * ( 10 ** -3 )         # in function
E_v_star = 29.8 * ( 10 ** -3 )      # in function

E_pi_list_1 = []    # in function
E_pi_list_2 = []    # in function
E_v_list = []       # in function
```

###If we interpret the figure on the 5th page of "An Off-Axis Neutrino Beam" (https://arxiv.org/pdf/hep-ex/0111033), as a plot of pion labframe energy, $E_{\pi}$, expressed a function of neutrino labframe energy, $E_{\nu}$.  As pointed out by my supervisor, such a consideration is logical if one notices that the plotted points are not evenly interspaced along the $E_{\pi}$ x-axis - suggesting that the function used to generate the plot took $E_{\nu}$ values as inputs.  Hence, unconvetionally, the y-axis represents its domain, and the x-axis its range.  

###A simple vertical (in this case, horizontal) line test reveals that the function must have a piecewise definition, which can be incorporated by considering principal and non-principal values of equation (1)

$(1) \;\;\;\;\;\;\;\;\;\;\;\;\;\; \cos(\theta^*)\;=\; \sqrt{1-\frac{E_{\nu}^2}{(E_{\nu}^{*})^2}\Bigl(\frac{1}{\cos^2(\theta)} - 1\Bigr) } \;\;\;,   $


###my function will contain both positive and negative values of the radical, taking in $E_{\nu}$ numbers (together with a predetermined $\theta$ value) and outputing $E_{\pi}$ values.

```python
######################## WORKING ON LOGIC TO BE INCORPORATED IN THE PLOTTING FUNCTION ########################

while len(E_pi_list_1) < 1000:

        E_v_list.append(E_v)
        cos_theta_star_1 =  (  ( 1 - ( ( E_v ** 2 ) / ( E_v_star ** 2 ) ) * ( ( 1 / ( ( cos(theta) ) ** 2 ) ) - 1 ) ) ** ( 1 / 2 ) )  ## Check this equation...
        E_pi_1_complex =  (  (m_pi * E_v )    /  (   E_v_star * ( 1 + cos_theta_star_1 )  )   )
        E_pi_1_real = E_pi_1_complex.real

        E_v = E_v + 0.001

        E_pi_list_1.append(E_pi_1_real)


E_v_list_reversed = list(reversed(E_v_list))


for ele in E_v_list_reversed:

    cos_theta_star_2 = - ( ( 1 - ( ( ele ** 2 ) / ( E_v_star ** 2 ) ) * ( ( 1 / ( ( cos(theta) ) ** 2 ) ) - 1 ) ) ** ( 1 / 2 ) )  ## Check this equation...

    if ( 1 + cos_theta_star_2) == 0:

        print("Error: division by zero for the ele: {ele}, cos_theta_star_2: {cos_theta_star_2}")

        continue # Skip to the next iteration

    E_pi_2_complex =  (  (m_pi * E_v )    /  (   E_v_star * ( 1 + cos_theta_star_2 )  )   )
    E_pi_2_real = E_pi_2_complex.real
    E_pi_list_2.append(E_pi_2_real)

del E_v_list_reversed[0]
```

```python
plt.scatter( E_pi_list_1 , E_v_list , s = 0.2 , c = "r" , marker = "x" )

plt.scatter( E_pi_list_2 , E_v_list_reversed , s = 0.2 , c = "b" , marker = "x" )

plt.title(r"Figure 1: $E_{\pi}$ against $E_{\nu}$")
plt.xlabel(r"$E_{\pi}$ (GeV)")
plt.ylabel(r"$E_{\nu}$ (GeV)")
plt.xlim(0, 15)
plt.show()
```

```python
########################## FIRST TRY ##########################
def Pion_to_Neutrino_converter( theta):
        E_pi_list_1 = []
        E_pi_list_2 = []
        E_v_list = []
        m_pi = 139.6 * ( 10 ** -3 )
        E_v_star = 29.8 * ( 10 ** -3 )
        E_pi_1_complex = 0
        E_pi_1_real = 0
        E_pi_2_complex = 0
        E_pi_2_real = 0
        E_v = 0

        while len(E_pi_list_1) < 1000:      ## There is a problem in the condition

                E_v_list.append(E_v)
                cos_theta_star_1 =  (  ( 1 - ( ( E_v ** 2 ) / ( E_v_star ** 2 ) ) * ( ( 1 / ( ( cos(theta) ) ** 2 ) ) - 1 ) ) ** ( 1 / 2 ) )  ## Check this equation...
                E_pi_1_complex =  (  (m_pi * E_v )    /  (   E_v_star * ( 1 + cos_theta_star_1 )  )   )
                E_pi_1_real = E_pi_1_complex.real

                E_v = E_v + 0.001

                E_pi_list_1.append(E_pi_1_real)


        E_v_list_reversed = list(reversed(E_v_list))


        for ele in E_v_list_reversed:

            cos_theta_star_2 = - ( ( 1 - ( ( ele ** 2 ) / ( E_v_star ** 2 ) ) * ( ( 1 / ( ( cos(theta) ) ** 2 ) ) - 1 ) ) ** ( 1 / 2 ) )  ## Check this equation...

            if ( 1 + cos_theta_star_2) == 0:

            #    print("Error: division by zero for the ele: {ele}, cos_theta_star_2: {cos_theta_star_2}")

                continue # Skip to the next iteration

            E_pi_2_complex =  (  (m_pi * E_v )    /  (   E_v_star * ( 1 + cos_theta_star_2 )  )   )
            E_pi_2_real = E_pi_2_complex.real
            E_pi_list_2.append(E_pi_2_real)

        del E_v_list_reversed[0]

        return E_pi_list_1 , E_pi_list_2 , E_v_list , E_v_list_reversed
```

###Figures 4 and 5 are indicative of a generated $E_{\nu}$ list of insufficient length.  Therefore, the condition of the while loop must be adjusted.  We require a condition that is sensitive to the gradient of the line, a condition that is no longer satisfied when the gradient vanishes.  This forms the "switch" that causes the function to employ the non-principal value of the aforementioned radical.

#Adjusting the while loop condition

```python
########################## SECOND TRY ##########################

################### There is a problem when theta == 0 #####################
def Pion_to_Neutrino_converter_SECOND_TRY( theta):
        E_pi_list_1 = []
        E_pi_list_2 = []
        E_v_list = []
        m_pi = 139.6 * ( 10 ** -3 )
        E_v_star = 29.8 * ( 10 ** -3 )
        E_pi_1_complex = 0
        E_pi_1_real = 0
        E_pi_2_complex = 0
        E_pi_2_real = 0
        E_pi_1_real_B = 0
        E_v = 0

        h = 10


        while h > 0.000001:


                E_v_list.append(E_v)

                if theta == 0:

                    cos_theta_star_1 = 1
                else:

                  cos_theta_star_1 =  (  ( 1 - ( ( E_v ** 2 ) / ( E_v_star ** 2 ) ) * ( ( 1 / ( ( cos(theta) ) ** 2 ) ) - 1 ) ) ** ( 1 / 2 ) )  ## There is an issue in this line when theta = 0

                E_pi_1_complex =  (  (m_pi * E_v )    /  (   E_v_star * ( 1 + cos_theta_star_1 )  )   )
                E_pi_1_real_B = E_pi_1_complex.real

                if E_pi_1_real_B != 0:
                    h = E_pi_1_real_B - E_pi_1_real



                E_pi_1_real = E_pi_1_real_B

                E_v = E_v + 0.001

                E_pi_list_1.append(E_pi_1_real)


        E_v_list_reversed = list(reversed(E_v_list))


        for ele in E_v_list_reversed:

            if theta == 0:

                    cos_theta_star_2 = -1
            else:

                    cos_theta_star_2 = - ( ( 1 - ( ( ele ** 2 ) / ( E_v_star ** 2 ) ) * ( ( 1 / ( ( cos(theta) ) ** 2 ) ) - 1 ) ) ** ( 1 / 2 ) )    ## There is an issue in this line when theta = 0



            if ( 1 + cos_theta_star_2) == 0:

            #    print("Error: division by zero for the ele: {ele}, cos_theta_star_2: {cos_theta_star_2}")

                continue # Skip to the next iteration

            E_pi_2_complex =  (  (m_pi * E_v )    /  (   E_v_star * ( 1 + cos_theta_star_2 )  )   )
            E_pi_2_real = E_pi_2_complex.real
            E_pi_list_2.append(E_pi_2_real)

        del E_v_list_reversed[0]

        return E_pi_list_1 , E_pi_list_2 , E_v_list , E_v_list_reversed
```

##This is the relative neutrino flux (Rel_Neu_Flux)

```python
cos_theta_star = (  ( 1 - ( ( E_v ** 2 ) / ( E_v_star ** 2 ) ) * ( ( 1 / ( ( cos(theta) ) ** 2 ) ) - 1 ) ) ** ( 1 / 2 ) )

E_pi = (  (m_pi * E_v )    /  (   E_v_star * ( 1 + cos_theta_star )  )   )

Rel_Neu_Flux = ( ( ( E_pro - E_pi ) ** 5 ) * ( ( E_pi * E_v ) / ( cos_theta_star ) ) )


```
